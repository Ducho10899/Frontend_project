```python
import dash
import dash_bootstrap_components as dbc
from dash import dcc, html, dash_table, Dash
from dash.dependencies import Output, Input
from collections import OrderedDict

import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.io as pio
pio.templates

import pandas as pd
import datetime
```

```python
df_market_trends = pd.read_excel('MarketTrends_Standard_National_2022-06-10.xlsx', sheet_name='Market Trends (Suburbs)')
df_market_trends = df_market_trends[df_market_trends['State'] == 'VIC']

## Create new dataframes by Property Type H and U
df_market_trends_H = df_market_trends.loc[(df_market_trends['Property Type'] == 'H')].reset_index(drop=True)
df_market_trends_U = df_market_trends.loc[(df_market_trends['Property Type'] == 'U')].reset_index(drop=True)
```

```python
df_percentage_growth_rate_H = df_market_trends_H[['Suburb','20 year annual compound growth rate in median sales price (12 months)', 
                                                          '10 year annual compound growth rate in median sales price (12 months)',
                                                          '5 year annual compound growth rate in median sales price (12 months)']].copy()
df_percentage_growth_rate_H = df_percentage_growth_rate_H.dropna()

df_percentage_growth_rate_H = df_percentage_growth_rate_H.rename(columns= {'5 year annual compound growth rate in median sales price (12 months)': '5 years', 
                                                            '10 year annual compound growth rate in median sales price (12 months)': '10 years',
                                                            '20 year annual compound growth rate in median sales price (12 months)': '20 years'})

df_percentage_growth_rate_H[['5 years', '10 years', '20 years']] = round(df_percentage_growth_rate_H[['5 years', '10 years', '20 years']]*100, 2)


# df_percentage_dwelling_type_H = df_percentage_dwelling_type_H.reset_index(drop="index")
df_percentage_growth_rate_H = df_percentage_growth_rate_H.melt(id_vars=['Suburb'], var_name ='Years', value_name ='Percentage')
df_percentage_growth_rate_H

import plotly.graph_objects as go

night_colors = ['rgb(56, 75, 126)', 'rgb(18, 36, 37)', 'rgb(34, 53, 101)',
                'rgb(36, 55, 57)', 'rgb(6, 4, 4)']

x = df_percentage_growth_rate_H[df_percentage_growth_rate_H['Suburb'] == 'Abbotsford'].Years
y = df_percentage_growth_rate_H[df_percentage_growth_rate_H['Suburb'] == 'Abbotsford'].Percentage

fig = px.bar(df_percentage_growth_rate_H, x=y, y=x, orientation='h',text_auto=True)
fig.update_traces(marker_color=night_colors, opacity=0.6)
fig.show()
```
